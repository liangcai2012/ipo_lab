# About Google Colab 

Google Colab is an online version of jupyter notebook, mainly for collaboration on machine learning algorithm education and learning.  This notebook is stored on Google driver and shared with other users via Google sharing. 


A notebook is a mix of text cells and python code snippet (ipython) cells.  The text make uses of markdown (a popular HTML markup language). But little or no knowledge of markdown is needed when editing text cells with embedded editor.  


The ipython code snippet executes in an environment Google provided to each user, so you might not be able to run the snippet on a shared notebook before you install all dependencies and obtained the data.  Ipython has three types of instructions:
* python code. 
* shell commands: started with !
* Magic commands: started with % or %%

All data used in this note book is obtained through github. It is always recommended to sync with the github repo each time open the notebook.
